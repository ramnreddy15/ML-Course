```python
import numpy as np
import math

file = open("./Advertising.csv")
csvFile = np.loadtxt(file, delimiter=",", skiprows=1)
n = csvFile.shape[0]
A = np.ones((n,4))
Y = csvFile[:,4]
A[:,0] = csvFile[:,1]
A[:,1] = csvFile[:,2]
A[:,2] = csvFile[:,3]
B = np.ones((4,1))
Q, R = np.linalg.qr(A, mode='complete')
solveY = np.dot(Q.T,Y)[:4]
vals = []
for arr in R:
    temp = []
    for val in arr:
        if val != 0:
            temp.append(val)
    vals = temp + vals

B[3] = solveY[3]/vals[0]
B[2] = (solveY[2]-vals[2]*B[3])/vals[1]
B[1] = (solveY[1]-vals[5]*B[3]-vals[4]*B[2])/vals[3]
B[0] = (solveY[0]-vals[9]*B[3]-vals[8]*B[2]-vals[7]*B[1])/vals[6]

print("Slope for TV:",B[0,0])
print("Slope for radio:",B[1,0])
print("Slope for newspaper:",B[2,0])

YF = A.dot(B)

RSS, TSS, yAvg, xAvg1, xAvg2, xAvg3 = 0, 0, 0, 0, 0, 0
TSS1, TSS2, TSS3 = 0,0,0
for y in range(n):
    yAvg+=YF[y,0]
    xAvg1+=A[y,0]
    xAvg2+=A[y,1]
    xAvg3+=A[y,2]
    
yAvg = yAvg/n
xAvg1 = xAvg1/n
xAvg2 = xAvg2/n
xAvg3 = xAvg3/n

for y in range (n):
    RSS += pow(YF[y,0]-Y[y],2)
    TSS += pow(yAvg-Y[y],2)
    TSS1 += pow(xAvg1-A[y,0],2)
    TSS2 += pow(xAvg2-A[y,1],2)
    TSS3 += pow(xAvg3-A[y,2],2)
    
RSE = math.sqrt(RSS/(n-4))
SE1 = RSE/math.sqrt(TSS1)
SE2 = RSE/math.sqrt(TSS2)
SE3 = RSE/math.sqrt(TSS3)
print("R^2:", 1 - (RSS/TSS))
print("RSE:", RSE)
print("SE1:", SE1)
print("SE2:", SE2)
print("SE3:", SE3)
print("t1:",(B[0]/(RSE/math.sqrt(TSS1)))[0])
print("t2:",(B[1]/(RSE/math.sqrt(TSS2)))[0])
print("t3:",(B[2]/(RSE/math.sqrt(TSS3)))[0])

from scipy import stats

print( 'p1:' , stats . t . sf( abs((B[0]/(RSE/math.sqrt(TSS1)))[0]) , df=n-4 ) * 2.0 )
print( 'p2:' , stats . t . sf( abs((B[1]/(RSE/math.sqrt(TSS2)))[0]) , df=n-4 ) * 2.0 )
print( 'p3:' , stats . t . sf( abs((B[2]/(RSE/math.sqrt(TSS3)))[0]) , df=n-4 ) * 2.0 )

print("TV interval: [", (B[0]-2*SE1)[0] , ",", (B[0]+2*SE1)[0], "]")
print("Radio interval: [", (B[1]-2*SE2)[0] , ",", (B[1]+2*SE2)[0], "]")
print("Newspaper interval: [", (B[2]-2*SE3)[0] , ",", (B[2]+2*SE3)[0], "]")
```
