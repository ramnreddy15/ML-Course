```python
import numpy as np
file = open("./test.csv")
csvFile = np.loadtxt(file, delimiter=",", skiprows=0)
n = csvFile.shape[0]
A = np.ones((n,2))
A[:,0] = csvFile[:,0]
B = csvFile[:,1]

M = csvFile[:,0:2]
R = np.corrcoef(M, rowvar = False)[0,1]
print("R^2:",R*R)

Q, R = np.linalg.qr(A, mode='complete')
QTB = (Q.T.dot(B))
B0 =QTB[1]/R[1,1]
B1 = (QTB[0]-R[0,1]*B0)/R[0,0]
print("B0:",B0)
print("B1:",B1)
```

```python
xp = np.linspace(-3,5,num=100) #B0
yp = np.linspace(-3,5,num=100) #B1
```

```python
X, Y = np.meshgrid(xp,yp)
z=np.zeros((yp.size,xp.size))
```

```python
RSS = 0
smallestRSS = 1000000000
rX = 0
rY = 0
for j in range(len(xp)):
    for k in range(len(yp)):
        for i in range(n):
            RSS = RSS+pow((csvFile[i][1]-(xp[j]+yp[k]*csvFile[i][0])),2)
        z[k][j] = RSS
        if RSS < smallestRSS:
            rX = j
            rY = k
            smallestRSS = RSS
        RSS = 0
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax = plt.axes(projection ="3d")
ax.plot_surface(X, Y, z)
ax.scatter(xp[rX], yp[rY], z[rY][rX],color="#ff0000",s=100)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
# ax = fig.add_subplot(111, projection='3d')
obj = ax.contour(X, Y, z)
ax.scatter(xp[rX], yp[rY],color="#ff0000",s=100)
ax . clabel( obj , obj . levels , inline = True , fmt = '%0.1f' )
ax . annotate( ' %0.1f' % z[rY][rX], (xp[rX],yp[rY]) )
plt.show()
```
