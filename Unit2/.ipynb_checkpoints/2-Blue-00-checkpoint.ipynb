```python
from sklearn.linear_model import LogisticRegression
#
# seed is 0
#

import numpy as np
import csv
X = []
Y = []
with open('Default.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    next(reader)
    for row in reader:
        X.append(float(row[2]))
        Y.append(1 if row[0] == '"Yes"' else 0)
        
X = np.asarray(X)
Y = np.asarray(Y)
X = X . reshape( -1 , 1 )
print( X . shape )
print( Y . shape )
clf = LogisticRegression( random_state = 0 ) . fit( X , Y )
print( clf . score( X , Y ) ) # mean accuracy
print( clf . intercept_ , clf . coef_ )
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
import math
print("Max balance is:",max(X)[0])
intercept = clf.intercept_[0]
coef = clf.coef_[0][0]
def doEquation(x):
    return pow(math.e,intercept+coef*x)/(1+pow(math.e,intercept+coef*x))

x = np.linspace(0, 3000, 10000)
y = [doEquation(xT) for xT in x]
plt.plot(x,y, color='C0', label='Regression Curve')
ax.legend()
ax.axis('auto')
plt.show()
print("Total Observations is:",len(X))
```

```python
def getPredVal(x):
    other_val = intercept + coef*x
    val = np.exp(other_val)/(1+np.exp(other_val))
    if val >= 0.5:
        return 1
    else:
        return 0

predY = [getPredVal(i) for i in X]
wrongX = []
wrongY = []
falsePos = 0
falseNeg = 0
for i,j in enumerate(predY):
    if(j!=Y[i]):
        if(j==0):
            falseNeg = falseNeg+1
        elif(j==1):
            falsePos = falsePos+1
        wrongY.append([1,0][j])
        wrongX.append(X[i])
plt.scatter(x , y , color='blue', s = 20 , linewidths = 1, marker='o', label="0")
plt.scatter(wrongX , wrongY , color='red', s = 1 , linewidths = 1, marker='o', label="0")
print("Total False Positives:", falsePos)
print("Total False Negatives:", falseNeg)
print("Total Observations in Dataset:", len(X))
```
