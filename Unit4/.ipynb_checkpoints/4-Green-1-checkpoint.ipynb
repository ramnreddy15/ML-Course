```python
import math

import numpy as np
import csv
X = []
Y = []
with open('faithful.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    next(reader)
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
```

```python
muX = sum(X)/len(X)
muY = sum(Y)/len(Y)
sigmaX = math.sqrt(sum([pow(muX - x,2) for x in X])/(len(X)-1))
sigmaY = math.sqrt(sum([pow(muY - y,2) for y in Y])/(len(Y)-1))
```

```python
transX = [(x-muX)/sigmaX for x in X]
transY = [(y-muY)/sigmaY for y in Y]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x1,x2,y1,y2 = [-1], [1], [1], [-1]
for i in range(len(X)):
    dist1 = pow(transX[i]+1,2) + pow(transY[i]-1,2)
    dist2 = pow(transX[i]-1,2) + pow(transY[i]+1,2)
    if(dist1<dist2):
        x1.append(transX[i])
        y1.append(transY[i])
    else:
        x2.append(transX[i])
        y2.append(transY[i])

plt.scatter(x1,y1, color='C0', label='Center 1')
plt.scatter(x2,y2, color='C1', label='Center 2')
ax.legend()
ax.axis('auto')
plt.show()
```
